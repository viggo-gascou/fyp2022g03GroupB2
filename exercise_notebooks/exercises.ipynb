```
import regex as re
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from functools import reduce
import difflib
```

```
with open("../doll_house_full_text.txt", "r") as f:
    doll_text = f.read()

with open("../tweeteval/datasets/irony/train_text.txt", "r") as f:
    irony = f.read()

with open("../tweeteval/datasets/stance/climate/train_text.txt", "r") as f:
    stance_climate = f.read()
```

```
with open("../pattern.txt", "r") as f:
    pat = re.compile(f.read())
```

```
#https://neptune.ai/blog/tokenization-in-nlp 
```

```
corp_book = re.findall(pat, doll_text)

corp_irony = re.findall(pat, irony)

corp_stance = re.findall(pat, stance_climate)
```

```
voc_book = Counter(corp_book)

voc_irony = Counter(corp_irony)

voc_stance = Counter(corp_stance)

#voc = dict(sorted(voc.items(), key=lambda item: item[1], reverse=True))
print(f'20 most common types in doll hosuse text: {voc_book.most_common(20)}')
print(f'20 most common types in irony text: {voc_irony.most_common(20)}')
print(f'20 most common types in stance climate text: {voc_stance.most_common(20)}')

def common_keys(dic, n):
    common = dic.most_common(n)
    return set(map(lambda x: x[0], common))
reduce(lambda x, y: x.intersection(y), list(map(lambda x: common_keys(x, 20), [voc_book, voc_irony, voc_stance])))
```

```
def zipf(corpus):
    voc = Counter(corpus)
    frq = pd.DataFrame(voc.most_common(), columns=['token', 'frequency'])

    # Index in the sorted list
    frq['idx'] = frq.index + 1

    # Frequency normalised by corpus size
    frq['norm_freq'] = frq.frequency / len(corpus)

    # Cumulative normalised frequency
    frq['cumul_frq'] = frq.norm_freq.cumsum()

    sns.set_theme(style='whitegrid')

    # Plot: Cumulative frequency by index
    sns.relplot(x='idx', y='cumul_frq', data=frq)
    plt.show()

    # Plot: Cumulative frequency by index, top 10000 tokens
    sns.relplot(x='idx', y='cumul_frq', data=frq[:10000], kind='line')
    plt.show()

    # Plot: Log-log plot for Zipf's law
    frq['log_frq'] = np.log(frq.frequency)
    frq['log_rank'] = np.log(frq.frequency.rank(ascending=False))
    sns.relplot(x='log_rank', y='log_frq', data=frq)
    plt.show()
```
